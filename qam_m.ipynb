# Параметры

```python
# Бит на символ
K = 2*3 # Должно быть кратно 2
# Кол-во символов QAM-M
M = (2**(K//2))**2
print "QAM-{}".format(M)
print "Бит на символ = {}".format(K)

# Длина входного сигнала в символах
L = 512
# Длина входного сигнала в битах
N = K*L
print "Длина входного сигнала = {}символ = {}бит".format(L, N)

# Символьная частота/скорость входного сигнала
FS = 100*1000
TS = 1.0/FS

print "Символьная частота входного сигнала = {}KГц".format(FS/1000)

# Длительность входного сигнала
DUR=L*TS
print "Длительность входного сигнала = {}мс".format(DUR*1000)

# Амплитуда модуляции
A = 10.0
print
print "Амплитуда модуляции = {}".format(A)

# Частота несущей
F0 = 100*1000
T0 = 1.0/F0
W0 = 2.0*math.pi*F0
print "Частота несущей = {}KГц".format(F0/1000)

"""
Кол-во отсчётов несущей на один символ входного сигнала.
Или кол-во отсчётов сигнала в канале на один символ.
"""
SCALE = 8
print "Кол-во отсчётов несущей на один символ = {}".format(SCALE)

# Расчетная (сырая) скорость. Реальная скорость меньше, в силу накладных расходов на служебную информацию
SPEED = K*FS
print
print "Расчетная скорость = {}Kбит/сек".format(SPEED/1000)

# Шаблонный сигнал, посылаемый для настройки приёмника в самом начале передачи
# Важно, чтобы модулированный шаблон имел максимально возможную мощность
TMPL_SYMB_0 = level_to_bit_seq(-A, K//2, A)
TMPL_SYMB_1 = level_to_bit_seq(A, K//2, A)
TMPL = numpy.concatenate([TMPL_SYMB_0, TMPL_SYMB_1, TMPL_SYMB_0, TMPL_SYMB_1, TMPL_SYMB_0, TMPL_SYMB_1, TMPL_SYMB_0, TMPL_SYMB_1])
TMPL_SYNCH_PART = numpy.array([-A, -A, -A, -A])
TMPL_QUAD_PART = -TMPL_SYNCH_PART
TMPL_L = len(TMPL)//K
print
print "Шаблонный сигнал:"
print TMPL

# Период следования синхро-последовательностей
INTER_SYNCH_PERIOD_IN_SYMBOLS = 16
#INTER_SYNCH_PERIOD_IN_SYMBOLS = None
# Кол-во символов в синхро-последовательности
SYNCH_SYMBOL_COUNT = 2
# Синхро-последовательность
SYNCH_TMPL = TMPL[:SYNCH_SYMBOL_COUNT*K]

# Разрядность ЦАП
DAC_N = 10
# Дин. диапазон в канале
MAX_LEVEL = 20.0
# Разрядность АЦП
ADC_N = 12
print
print "Разрядность ЦАП = {}бит".format(DAC_N)
print "Разрядность АЦП = {}бит".format(ADC_N)

# Длина кабеля
CABLE_LEN = 5*1000
print
print "Длина кабеля = {}км".format(CABLE_LEN/1000)

# Удельное затухание в кабеле Дб/м
ATTm = 0.004375
# Затухание Дб
ATT = CABLE_LEN*ATTm
print "Удельное затухание в кабеле = {}Дб/м".format(ATTm)
print "Затухание в кабеле = {}Дб".format(ATT)

# Максимальная частота полосы пропускания кабеля
#CABLE_CRITICAL_FREQ = 50*1000
CABLE_CRITICAL_FREQ = 200*1000
print "Максимальная частота полосы пропускания кабеля = {}КГц".format(CABLE_CRITICAL_FREQ/1000)

# Амплитуда шума
#NOISE_A = 0.1*A
NOISE_A = 0.01*A
print "Амплитуда шума = {}".format(NOISE_A)

# Дополнительное смещение приёмника относительно источника в долях отсчёта
INTERSAMPLE_RELATIVE_OFFSET = 0.5
print "Смещение приёмника относительно источника = {}отсчёта".format(INTERSAMPLE_RELATIVE_OFFSET)
```

# Входной сигнал

Cигнал состоит из *шаблона* и *рандомной* последовательности бит. Общая длина сигнала - *N бит*.

*Шаблон* - это заранее известная и фиксированная часть сигнала, необходимая для детектировани сигнала, корректировки мощности и фазы несущей. Первые len(TMPL) отсчётов - *шаблон*. Его перепады хорошо видны на графике. Шаблон подбирается таким образом, чтобы после модуляции максимизировать его мощность (вероятность прохождения по каналу) и упростить его детектироваие в приёмнике.

```python
# Генерация входного сигнала
if INTER_SYNCH_PERIOD_IN_SYMBOLS is None:
    signal = numpy.concatenate([TMPL, numpy.random.random_integers(0, 1, N-len(TMPL))])
else:
    signal = numpy.concatenate([TMPL, numpy.random.random_integers(0, 1, INTER_SYNCH_PERIOD_IN_SYMBOLS*K)])
    while len(signal) < N:
        signal = numpy.concatenate([signal, SYNCH_TMPL, numpy.random.random_integers(0, 1, INTER_SYNCH_PERIOD_IN_SYMBOLS*K)])
    signal = signal[:N]

# График входного сигнала
pyplot.plot(xrange(len(signal)), signal, "ro:")
pyplot.title("Input signal (bits)")
pyplot.xlabel("Bit number")
pyplot.xlim(0, 12*K)
pyplot.ylim(-0.1, 1.1)
pyplot.show()
```

# Разделение входного сигнала на синфазную и квадратурную части

Это предварительный этап перед непосредственной модуляцией сигнала. Последовательность бит входного сигнала необходимо разбить на две части для каждого из каналов (синфазного и квадратурного).
Для этого:
1. Исходная последовательность бит разбивается на подпоследовательности по *K/2 бит*. То есть последовательность бит [0,1,0,1,1,0,1,1,1,0,0,0] будет представлена как [[0,1,0], [1,1,0], [1,1,1], [0,0,0]], при K = 6.
2. В *синфазную* часть *уходят* все четные подпоследовательности. В *квадратурную* - все нечетные. То есть *синфазная* часть [[0,1,0], [1,1,1]], и *квадратурную* - [[1,1,0], [0,0,0]].
3. Каждая подпоследовательность бит преобразуется в уровень *от -A до A* при помощи функции *bit_seq_to_level* для случая прямоугольного созвездия (Constellation Diagram).

Получаем две последовательности уровней *от -A до A* длиной *L* (кол-во символов).

В случае прямоугольного созвездия и *некогерентного приёмника* для снижения уровня ошибок следует использовать *код Грея*. Таким образом, битовые представления ближайших уровней будут отличаться на один бит и возможная ошибка сдвига фазы несущей будет минимизирована.

```python
signal_parts = map(lambda x: bit_seq_to_level(x, A), signal.reshape(2*L, K//2))

# Синфазная часть
signal_synch_part = numpy.array(signal_parts[0:2*L:2])
# Квадратурная часть
signal_quad_part = numpy.array(signal_parts[1:2*L:2])

# График синфазной части
pyplot.subplot(2, 1, 1)
t = numpy.linspace(0, DUR, L, endpoint=False)
pyplot.plot(t*1000, signal_synch_part, "yo:")
pyplot.title("Input signal parts")
pyplot.xlim(0, min(L, 16)*TS*1000)
center_plot_axis()
pyplot.ylabel("Synch part")
pyplot.grid()

# График квадратурной части
pyplot.subplot(2, 1, 2)
pyplot.plot(t*1000, signal_quad_part, 'go:')
pyplot.xlim(0, min(L, 16)*TS*1000)
center_plot_axis()
pyplot.xlabel("Time (ms)")
pyplot.ylabel("Quad part")
pyplot.grid()

pyplot.show()
```

# Несущая

Данный график необходим для оценки адекватности выбранного параметра *SCALE* (кол-во отсчётов несущей на один символ). Слишком маленькие значения данного параметры приведут к потере точности при расчёте модулированного сигнала и потере информации. Слишком большие - увеличивают требования к ресурсам приёмника и передатчика.

```python
"""
Функция генерации несущей с учетом межсимвольного интервала
@symbol_count - кол-во символов
return - пару (отсчёты времени, отсчёты несущей)
"""
def carrier(symbol_count, phase=0.0):
    return synch.carrier(W0, symbol_count, TS, SCALE, phase)

# Несущая
carrier_cos = carrier(L, math.pi/2.0)
carrier_sin = carrier(L)

# График несущей
pyplot.plot(carrier_cos[0]*1000, carrier_cos[1], "r.:")
pyplot.title("Carrier")
pyplot.xlim(0, 2*T0*1000)
pyplot.ylim(-1.1, 1.1)
center_plot_axis()
pyplot.xlabel("Time (ms)")

pyplot.show()
```

# Constellation Diagram

Диаграмма показывает расположение символов *QAM-M* в комплексной области. Здесь представлен наиболее простой вариант расположения - *прямоугольный*. Символы занимают вершины сетки с равномерным шагом. То есть расстояние между двумя символами по любой из осей кратно определённой величине.

```python
def num_to_bit_seq(num, bit_count=K//2):
    return [(int(num)>>i)&1 for i in xrange(bit_count)]
levels = map(lambda x: bit_seq_to_level(num_to_bit_seq(x), A), xrange(2**(K//2)))

for level in levels:
    pyplot.plot(levels, [level]*len(levels), "rx")
pyplot.title("Constellation Diagram")
pyplot.xlim(-1.1*A, 1.1*A)
pyplot.ylim(-1.1*A, 1.1*A)
center_plot_axis()
pyplot.xlabel("I (synch)")
pyplot.ylabel("Q (quad)")
pyplot.minorticks_on()
pyplot.grid()

pyplot.show()
```

# ЦАП

```python
"""
Преобразование отсчёта в целое значение для ЦАП, а затем в уровень в канале
"""
def DAC(sample, max_sample):
    max_DAC = 2**(DAC_N-1) - 1
    return int(round(sample*max_DAC/max_sample))*MAX_LEVEL/max_DAC

# Максимальное абсолютное значение
max_sample = max(abs(mod_signal))
# Отсчёты для ЦАП
mod_signal_DAC = numpy.array(map(lambda sample: DAC(sample, max_sample), mod_signal))
```

# АЦП

```python
"""
Оцифровка принятого отсчёта
"""
def ADC(sample):
    max_ADC = 2**(ADC_N-1) - 1
    return int(round(sample*max_ADC/MAX_LEVEL))

# Оцифрованный сигнал на выходе из кабеля
mod_signal_after_cable_ADC = numpy.array(map(ADC, mod_signal_after_cable))
```

# 1-ый этап синхронизации

Детектирование начала пакета по методу FCR с шаблоном.

```python
# Диапазон поиска пакета в буфере. Максимальный номер отсчёта, до которого осуществлять поиск в буфере
FCR_DETECT_MAX_OFFSET = 8*SCALE
# Кол-во символов учавствующих в корректировке фазы
FCR_TEST_SYMBOL_COUNT = 2*TMPL_L
# Пороговое значение для среднеквадратической ошибки. Выбирается в соответствии с наименьшим расстоянием между симолами в созвездии.
DELTA_LEVEL = 2*A/(2**(K//2) - 1)
FCR_TH = 2*(DELTA_LEVEL**2)/4

print "FCR_TH = {}".format(FCR_TH)

"""
функция определения ближайшего уровня
"""
def closest_level(level):
    return bit_seq_to_level(level_to_bit_seq(level, K//2, A), A)

# Номер отсчёта mod_signal_after_cable_ADC, начиная с которого детектирован сигнал, и коэфф. усиления
detected_shift, optimal_gain = synch.fcr_detect(mod_signal=mod_signal_after_cable_ADC, max_offset=FCR_DETECT_MAX_OFFSET,
                            tmpl_synch_part=TMPL_SYNCH_PART, test_symbol_count=FCR_TEST_SYMBOL_COUNT,
                            w=W0, symbol_period=TS, scale=SCALE, closest_level=closest_level, th=FCR_TH)

if (detected_shift is None):
    raise Exception("Package cann't be detected")
print "Смещение пакета в буфере = {}".format(detected_shift)
print "Коэфф. усиления = {}".format(optimal_gain)

# Детектированный сигнал
detected_mod_signal = mod_signal_after_cable_ADC[detected_shift:]
# Кол-во целых символов в детектированном сигнале
detected_symbol_count = len(detected_mod_signal)//SCALE
print "Кол-во дектектированных сиволов = {}".format(detected_symbol_count)
# Детектированный и урезанный до кол-ва целых символов сигнал
detected_mod_signal = detected_mod_signal[:detected_symbol_count*SCALE]
```

# 2-ой этап синхронизации

Точная подстройка фазы по методу последовательных приближений с опорой на шаблон.

```python
# Точность корректировки фазы по ММП
MSA_EPS = pi/(16*180)

# Оптимальная фаза демодуляции, и коэфф. усиления
optimal_demod_phase, optimal_gain = synch.msa_synch(mod_signal=detected_mod_signal, tmpl_symbol_count=TMPL_L, max_mod_level=A,
                                                    w=W0, symbol_period=TS, scale=SCALE, eps=MSA_EPS, log=True)
    
print "Оптимальная фаза демодуляции = {}рад".format(optimal_demod_phase)

# График детектировнного сигнала
pyplot.plot(carrier_cos[0][:len(detected_mod_signal)]*1000, detected_mod_signal, "r")
pyplot.title("Detected mod-ed signal")
pyplot.xlim(0, min(L, 16)*TS*1000)
center_plot_axis()
pyplot.xlabel("Time (ms)")
pyplot.grid()

pyplot.show()
```

# Демодуляция с непрерывной корректировкой фазы по методу FCR

Демодуляция происходит, путём умножения полченного сигнала на несущую. На cos для получения синфазной части, и на sin для получения квадратурной части.

При демодуляции необходимо учитывать флуктуации фазы на протяжении пакета. Для этого, каждые FCR_CORRECTION_PERIOD_IN_SYMBOLS символов необходимо корректировать фазу по методу FCR.

```python
if INTER_SYNCH_PERIOD_IN_SYMBOLS is None:
    
    # Периодичность перекорректировки фазы по методу FCR
    FCR_CORRECTION_PERIOD_IN_SYMBOLS = FCR_TEST_SYMBOL_COUNT
    # Параметры окрестности поиска фазы по методу FCR
    FCR_PHASE_DELTA = 0.5*pi/SCALE
    FCR_PHASE_COUNT = int(round(FCR_PHASE_DELTA/MSA_EPS))
    
    best_phase = optimal_demod_phase
    best_gain = optimal_gain
    
    """
    Функция демодуляции фрагмента сигнала с корректировкой фазы по методу FCR
    """
    def demodulate_using_fcr(mod_signal_slice, slice_phase_shift):
        global best_phase, best_gain
        if ((mod_signal_slice is None) or (len(mod_signal_slice) == 0)):
            return None
        symbol_count = len(mod_signal_slice)//SCALE
        phases = numpy.concatenate([[slice_phase_shift + best_phase], slice_phase_shift + best_phase + numpy.linspace(
                                                                      -FCR_PHASE_DELTA/2, FCR_PHASE_DELTA/2, FCR_PHASE_COUNT)])
        # Поиск наилучшей фазы в окрестности оптимальной
        best_phase = synch.fcr_synch(mod_signal=mod_signal_slice, phases=phases, test_symbol_count=min(FCR_TEST_SYMBOL_COUNT, symbol_count),
                                     w=W0, symbol_period=TS, scale=SCALE, closest_level=closest_level,
                                     gain=best_gain) - slice_phase_shift
        print "* {}рад (gain {})".format(abs(optimal_demod_phase - best_phase), best_gain)
        # Пересчёт несущей с учетом оптимальной фазы демодуляции
        carrier_cos = carrier(symbol_count, slice_phase_shift + math.pi/2.0 + best_phase)
        carrier_sin = carrier(symbol_count, slice_phase_shift + best_phase)
        # Демодулированные части сигнала. См. пояснение по формуле демодуляции ниже.
        demod_signal_synch_part = best_gain*rcosflt_unscale_signal(2*mod_signal_slice*carrier_cos[1], SCALE)
        demod_signal_quad_part = best_gain*rcosflt_unscale_signal(2*mod_signal_slice*carrier_sin[1], SCALE)
        return (demod_signal_synch_part, demod_signal_quad_part)

    # Принятый сигнал разбивается на фрагменты по FCR_CORRECTION_PERIOD_IN_SYMBOLS символов
    # Кол-во целых фраментов
    full_slice_count = detected_symbol_count//FCR_CORRECTION_PERIOD_IN_SYMBOLS
    # Фрагменты принятого сигнала размером FCR_CORRECTION_PERIOD_IN_SYMBOLS символов
    detected_mod_signal_slices = detected_mod_signal[:full_slice_count*FCR_CORRECTION_PERIOD_IN_SYMBOLS*SCALE].reshape(
            full_slice_count, FCR_CORRECTION_PERIOD_IN_SYMBOLS*SCALE)

    # Части демодулированного сигнала
    demod_signal_synch_part = []
    demod_signal_quad_part = []

    # Фаза текущего сегмента
    slice_phase_shift = 0.0
    
    # Демодуляция целых фрагментов
    for signal_slice in detected_mod_signal_slices:
        demod_signal_slice_parts = demodulate_using_fcr(signal_slice, slice_phase_shift)
        demod_signal_synch_part.extend(demod_signal_slice_parts[0])
        demod_signal_quad_part.extend(demod_signal_slice_parts[1])
        slice_phase_shift = slice_phase_shift + FCR_CORRECTION_PERIOD_IN_SYMBOLS*SCALE*2*pi/SCALE

    # Размер последнего фрагмента
    last_slice_symbol_count = detected_symbol_count%FCR_CORRECTION_PERIOD_IN_SYMBOLS
    if (last_slice_symbol_count > 0):
        # Последний фрагмент
        detected_mod_signal_last_slice = detected_mod_signal[-last_slice_symbol_count*SCALE:]
        # Демодуляция последнего фрагмента
        last_slice_parts = demodulate_using_fcr(detected_mod_signal_last_slice, slice_phase_shift)
        demod_signal_synch_part.extend(last_slice_parts[0])
        demod_signal_quad_part.extend(last_slice_parts[1])

    demod_signal_synch_part = numpy.array(demod_signal_synch_part)
    demod_signal_quad_part = numpy.array(demod_signal_quad_part)
```

# Демодуляция с непрерывной корректировкой фазы по МПП

Непрерывная корректировка по МПП требует вставок шаблона с заданной периодичностью INTER_SYNCH_PERIOD_IN_SYMBOLS.

```python
if not(INTER_SYNCH_PERIOD_IN_SYMBOLS is None):
    
    # Периодичность перекорректировки фазы по МПП
    MSA_CORRECTION_PERIOD_IN_SYMBOLS = SYNCH_SYMBOL_COUNT+INTER_SYNCH_PERIOD_IN_SYMBOLS
    
    best_phase = optimal_demod_phase
    best_gain = optimal_gain
    
    """
    Функция демодуляции фрагмента сигнала с корректировкой фазы по МПП
    """
    def demodulate_using_msa(mod_signal_slice, slice_phase_shift, tmpl_symbol_count=SYNCH_SYMBOL_COUNT):
        global best_phase, best_gain
        if ((mod_signal_slice is None) or (len(mod_signal_slice) == 0)):
            return None
        symbol_count = len(mod_signal_slice)//SCALE
        # Поиск наилучшей фазы по МПП
        best_phase, best_gain = synch.msa_synch(mod_signal=mod_signal_slice, tmpl_symbol_count=tmpl_symbol_count, max_mod_level=A,
                                          w=W0, symbol_period=TS, scale=SCALE,
                                          phase_shift=slice_phase_shift, eps=MSA_EPS, log=False)
        print "* {}рад (gain {})".format(abs(optimal_demod_phase - best_phase), best_gain)
        # Пересчёт несущей с учетом оптимальной фазы демодуляции
        carrier_cos = carrier(symbol_count, slice_phase_shift + math.pi/2.0 + optimal_demod_phase)
        carrier_sin = carrier(symbol_count, slice_phase_shift + optimal_demod_phase)
        # Демодулированные части сигнала. См. пояснение по формуле демодуляции ниже.
        demod_signal_synch_part = optimal_gain*rcosflt_unscale_signal(2*mod_signal_slice*carrier_cos[1], SCALE)
        demod_signal_quad_part = optimal_gain*rcosflt_unscale_signal(2*mod_signal_slice*carrier_sin[1], SCALE)
        return (demod_signal_synch_part, demod_signal_quad_part)

    # Части демодулированного сигнала
    demod_signal_synch_part = []
    demod_signal_quad_part = []
    
    # Фаза текущего сегмента
    slice_phase_shift = 0.0
    
    # Демодуляция первого фрагмента с шаблоном
    first_slice_parts = demodulate_using_msa(detected_mod_signal[:(TMPL_L+INTER_SYNCH_PERIOD_IN_SYMBOLS)*SCALE],
                                             slice_phase_shift=slice_phase_shift, tmpl_symbol_count=TMPL_L)
    demod_signal_synch_part.extend(first_slice_parts[0])
    demod_signal_quad_part.extend(first_slice_parts[1])
    
    slice_phase_shift = slice_phase_shift + (TMPL_L+INTER_SYNCH_PERIOD_IN_SYMBOLS)*SCALE*2*pi/SCALE
    
    remaining_symbol_count = detected_symbol_count - (TMPL_L+INTER_SYNCH_PERIOD_IN_SYMBOLS)
    remaining_mod_signal = detected_mod_signal[(TMPL_L+INTER_SYNCH_PERIOD_IN_SYMBOLS)*SCALE:]
    
    # Принятый сигнал разбивается на фрагменты по MSA_CORRECTION_PERIOD_IN_SYMBOLS символов
    # Кол-во целых фраментов
    full_slice_count = remaining_symbol_count//MSA_CORRECTION_PERIOD_IN_SYMBOLS
    # Фрагменты принятого сигнала размером MSA_CORRECTION_PERIOD_IN_SYMBOLS символов
    detected_mod_signal_slices = remaining_mod_signal[:full_slice_count*MSA_CORRECTION_PERIOD_IN_SYMBOLS*SCALE].reshape(
            full_slice_count, MSA_CORRECTION_PERIOD_IN_SYMBOLS*SCALE)

    # Демодуляция целых фрагментов
    for signal_slice in detected_mod_signal_slices:
        demod_signal_slice_parts = demodulate_using_msa(signal_slice, slice_phase_shift)
        demod_signal_synch_part.extend(demod_signal_slice_parts[0])
        demod_signal_quad_part.extend(demod_signal_slice_parts[1])
        slice_phase_shift = slice_phase_shift + MSA_CORRECTION_PERIOD_IN_SYMBOLS*SCALE*2*pi/SCALE

    # Размер последнего фрагмента
    last_slice_symbol_count = remaining_symbol_count%MSA_CORRECTION_PERIOD_IN_SYMBOLS
    if (last_slice_symbol_count > 0):
        # Последний фрагмент
        remaining_mod_signal_last_slice = remaining_mod_signal[-last_slice_symbol_count*SCALE:]
        # Демодуляция последнего фрагмента
        last_slice_parts = demodulate_using_msa(remaining_mod_signal_last_slice, slice_phase_shift)
        demod_signal_synch_part.extend(last_slice_parts[0])
        demod_signal_quad_part.extend(last_slice_parts[1])

    demod_signal_synch_part = numpy.array(demod_signal_synch_part)
    demod_signal_quad_part = numpy.array(demod_signal_quad_part)
```
