# Корректирующие коды

# Введение

Так как шум и меняющиеся условия среды - это неотъемлема часть любого канала, необходимость реализации *корректирующих кодов* в данной задаче является очевидным этапом.

Были рассмотрены различные виды *линейных кодов*. Если одни лучше подходят для одиночных ошибок, то другие хороши для пакетов ошибок. Если одни лучше для каналов с белым шумом, то другие хороши в случае систематической ошибки. Существенно разнятся и требования к ресурсам.

Помимо этого, если считать ресурс (приёмника и передатчика) неограниченным, то правильным выбором будет семейство *каскадных кодов*. *Каскадный код* сочетает в себе несколько подходов (пример - циклический код Рида-Соломона и сверточный код), аккумулируя их достоинства.

Были выделены следующие стратегии защиты от ошибок:

1. Обнаружение ошибок и повтор передачи.
2. Обнаружение ошибок и отбраковка пакета.
3. Корректировка ошибок.

Любой корректирующий код существенно эффективнее детектирует ошибки, нежели их корректирует. Дургими словами, после корректного детектирования ошибки, может последовать некорректная корректировка, приводящая к большему числу ошибок. В связи с этим и неопределённостью модели среды, адекватным выбором будет наиболее надежная - первая стратегия. Для целей обнаружения ошибок, в большинстве случаев, хорошо подходит *CRC* - циклический избыточный код (подсемейство линейных кодов).

Также широкое применение находит свёрточное кодирование с целью увеличить соотношение сигнал/шум. В сочетании с контролем ошибок на основе CRC - является отличным *стартовым* решением для большинства задач.

# CRC (Циклический избыточный код)

Служит для обнаружения ошибок. Реже - для корректировки. Прост в реализации и не требователен к ресурсам. Далее приводится универсальная табличная реализация.

```python
"""
Универсальная реализация CRC
с порождающим многочленом степени кратной 8
(CRC8, CRC16, CRC32, CRC64, ...)
"""
class CRC:
    """
    Конструктор
    @width - степень порождающего многочлена (разрядность регистра)
    @poly - порождающий полином
    @initial - начальное состояние регистра
    @xor_out - число, с которым складывается по модулю 2 результат
    """
    def __init__(self, width, poly, initial, xor_out=0):
        self.__width = width
        self.__reg_mask = (1<<width)-1
        self.__poly = poly
        self.__initial = initial
        self.__xor_out = xor_out
        self.__table = self.__create_table()
    
    """
    Табличный алгоритм генерации CRC
    @bit_seq - последовательность бит
    return - CRC соответствующий последовательности
    """
    def gen(self, bit_seq):
        reg = self.__initial
        # Преобразование последовательности бит в последовательность байт
        byte_seq = numpy.packbits(bit_seq.reshape(len(bit_seq)//8, 8))
        # Обработка каждого байта
        for byte in byte_seq:
            reg = self.__table[(reg^byte)&0xFF] ^ (reg>>8);
        return reg^self.__xor_out
    
    """
    @bit_seq - последовательность бит
    return - защищенная CRC последовательность бит
    """
    def protect(self, bit_seq):
        code = number_to_bit_seq(self.gen(bit_seq), self.__width, False)
        return numpy.concatenate([code, bit_seq])
    
    def check(self, protected_bit_seq):
        code = protected_bit_seq[0:self.__width]
        bit_seq = protected_bit_seq[self.__width:]
        return (self.gen(bit_seq) == bit_seq_to_number(code[::-1]))
    
    """
    Гнереация таблицы CRC кодов для состояний от 0 до 255 (byte)
    """
    def __create_table(self):
        return numpy.array(map(lambda i: self.__eight_cycles(i), xrange(256)), dtype=long)
    
    def __eight_cycles(self, i):
        reg = i
        for k in xrange(8):
            reg = (reg>>1)^self.__poly[1] if reg&1 else reg>>1
            reg = reg&self.__reg_mask
        return reg

    """
    Вывод общей информации о данной разновидности CRC в строку
    """
    def __str__(self):
        return "CRC{}({:02X})".format(self.__width, self.__poly[0])
```

## Стандартные CRC

```python
CRC16 = CRC(16, (0x8005, 0xA001), 0x0000)
CRC16_IBM = CRC(16, (0x8005, 0xA001), 0xFFFF)
CRC32 = CRC(32, (0x04C11DB7, 0xEDB88320), 0xFFFFFFFF, 0xFFFFFFFF)

# Тестовая последовательность (из спецификации)
test_bit_seq = string_to_bit_seq("123456789")

# Проверка корректности работы по спецификации
assert (0xBB3D == CRC16.gen(test_bit_seq)), "CRC16 check failed"
assert (0x4B37 == CRC16_IBM.gen(test_bit_seq)), "CRC16_IBM check failed"
assert (0xCBF43926 == CRC32.gen(test_bit_seq)), "CRC32 check failed"

print "OK"
```

## Тест CRC

```python
# Длина сигнала
N = 512 
# Случайный сигнал
signal = numpy.random.random_integers(0, 1, N)

# Сигнал дополненный CRC кодом
protected_sinal = CRC32.protect(signal)

# Проверка защищенного сигнала на отсутствие ошибок (проверка должна быть пройдена)
assert CRC32.check(protected_sinal), "Something wrong with protect/check #1"

# Вектор ошибок
error = numpy.random.random_integers(0, 1, len(protected_sinal))
# "Внедрение" ошибок в защищенный сигнал
signal_with_error = numpy.bitwise_xor(protected_sinal, error)

# Проверка защищенного сигнала на наличие ошибок (проверка не должна быть пройдена)
assert not CRC32.check(signal_with_error), "Something wrong with protect/check #2"

print "OK"
```

# Convolution code (Свёрточный код)

*Свёрточный код* часто используется *для корректировки ошибок* на канальном уровне. В отличие от других линейных кодов при *свёрточном кодировании* исходный сигнал воспринимается как *поток бит*. Выход *свёрточного кодера* определяется как линейная функция от *"хвоста" сигнала* (последние несколько бит) и *внутреннего состояния*, которое в свою очередь опеределется предыдущими итерациями работы кодера. Другими словами - это свёртка сигнала с участием линейной функции.

*Сильные стороны* свёрточных кодов:
* Простота реализации и низкие требования к ресурсам кодера - сдвиговый регистр и несколько сумматоров по модулю 2.
* Хорошие показатели в каналах с белым шумом.

![Свёрточный кодер](images/conv-code-1-2.gif)

*Недостатки* свёрточных кодов: 
* Главным недостатком является *распространение ошибки* и "превращение" её в *пакет ошибок* при невозможности корректировки. Данный недостаток компенсируется на пакетном уровне контролем через *CRC*.
* Быстрый рост (степенная функция) сложности алгоритма декодера при увеличении ширины кодера. Поэтому, как правило, ширина берётся равной 1-7 битам.

В общем случае *свёрточный код* работает с потоком последовательностей бит. Длина элементарной последовательности бит - это ширина (входа) *свёрточного кодера*.

Декодирование *свёрточного кода* - это поиск наиболее правдоподобного пути в пространстве состояний кодера. В общем случае требует перебора всех состояний. Аналогично задачам поиска в пространстве решений, оптимизация достигается удачным подбором *эвристик* (целевых функций), "направляющих" функцию поиска. То есть, экономия ресурсов за счёт снижения веротяности найти правильное решение. Алгоритм декодирования *Витерби* относится к эвристическим.

# Обобщенный свёрточный кодер/декодер

Данная реализация отталкивается от *общей идеи*. На её основе могут быть построены различные алгоритмы свёрточного кодирования/декодирования. В частоности - *декодер Витерби*.

Для управления производительностью и *кодовым расстоянием* введён дополнительный параметр - *глубина (де)кодирования* (depth). После *depth* итераций работы (де)кодера происходит выбор наилучшего пути и сброс внутреннего состояния. Слишком большая *глубина (де)кодирования* увеличивает сложность алгоритма и уменьшает кодовое расстояние, снижая вероятность корректировки сигнала. Слишком маленькая - вырождает алгоритм, превращая его в *жадный*. Значения глубины кодирования и глубины декодирования должны совпадать.

```python
# Максимально возможное кол-во выживших на глубине декодирования путей. Влияет на производительность алгоритма.
MAX_SURVIVED_PATH_COUNT = 16

"""
Простейшая функция фильтрации списка вершин вероятных путей по жадному алгоритму. Неустойчива к ошибкам.
@path_heads - вершины вероятных путей
return - список вершин "выживших" путей
"""
def greedy_path_filter(path_heads):
    if ((path_heads is None) or (len(path_heads) == 0)): return numpy.array([], dtype=object)
    # Всевозможные расстояния Хэмминга для данного списка вершин путей
    distances = map(lambda path_head: path_head.dist(), list(path_heads))
    # Наилучшая/минимальная дистанция
    min_dist = min(distances)
    # Список вершин путей с наилучшей/минимальной дистанцией
    return filter(lambda path_head: path_head.dist() == min_dist, path_heads)

"""
Вспомогательная функция выбирающая произвольный элемент последовательности
@seq - массив, список, последовательность
return 
"""
def arbitrary_one(seq):
    return None if ((seq is None) or (len(seq) == 0)) else numpy.random.choice(seq)

"""
Вспомогательная функция избирающая count произвольных элементов последовательности
@seq - массив, список, последовательность
return - список из count произвольных элементов @seq
"""
def arbitrary_elements(seq, count):
    if ((seq is None) or (len(seq) == 0)):
        return numpy.array([])
    elif (len(seq) <= count):
        return seq
    else:
        return numpy.random.choice(seq, count, False)

"""
Универсальная реализация свёрточного кодера
"""
class CC:
    """
    Конструктор
    @k - ширина входа кодера
    @m - кол-во ячеек внутреннего регистра (> 1)
    @gs - список порождающих полиномов, на каждый из выходов кодера
    """
    def __init__(self, k, m, gs):
        self.__k = k
        # Вход должен иметь ширину
        assert k > 0
        self.__m = m
        # Внутренняя память должна состоять более чем из одной ячейки
        assert m > 1
        self.__reg_mask = (1<<(k*m))-1
        self.__gs = gs
        # Выход должен быть шире входа, так как избыточен
        assert len(gs) > k

    """
    Функция свёрточного кодирования
    @bit_seq - последовательность бит
    @depth - глубина (де)кодирования - кол-во итераций, после которых происходит сброс внутреннего состояния
    return - закодированная последовательность бит
    """
    def encode(self, bit_seq, depth=8):
        if ((bit_seq is None) or (len(bit_seq) == 0)): return numpy.array([], dtype=int)
        result = []
        # Кол-во элементарных подпоследовтельностей в @bit_seq
        count = len(bit_seq)//self.__k
        # Элементарные подпоследовательности @bit_seq
        bit_seq_parts = bit_seq.reshape(count, self.__k)
        # Регистр состояния кодера
        reg = 0
        # Цикл свёрточного кодирования
        for i in xrange(count):
            # Если достигнута глубина декодирования, то сброс внутреннего состояния
            if (i%depth == 0): reg = 0
            # Элементарная подпоследовательность из @bit_seq как число
            part_num = bit_seq_to_number(bit_seq_parts[i])
            (out_part, reg) = self.one_cycle_f(part_num, reg)
            result.extend(out_part)
            
        return numpy.array(result, dtype=int)
    
    """
    Функция одного цикла работы кодера
    @part_num - элементарная подпоследовательность (как целое число)
    @reg - состояние кодера
    return - пара (закодированная элементарная подпоследовательность, новое состояние кодера)
    """
    def one_cycle_f(self, part_num, reg):
        # Новое состояние кодера
        next_reg = self.next_reg(part_num, reg)
        # Закодированная элементарная подпоследовательность
        out_part = map(lambda g: xor_bits(next_reg&g), self.__gs)
        return (out_part, next_reg)
    
    """
    Вспомогательная функция, вычисляющая следующее состояние кодера из текущего
    @part_num - элементарная подпоследовательность (как целое число)
    @reg - состояние кодера
    return - следующее состояние кодера
    """
    def next_reg(self, part_num, reg):
        return ((reg<<self.__k)|part_num)&self.__reg_mask
        
    """
    Обратная к @one_cycle_f функция.
    Может быть реализована как табличная функция.
    @target_out_part - закодированная элементарная подпоследовательность (как последовательность бит)
    @reg - состояние кодера
    return - исходная элементарная подпоследовательность (как число)
    """
    def rev_one_cycle_f(self, target_out_part, reg):
        # Все возможные значения входной элементарной подпоследовательности
        possible_parts = xrange(1<<self.__k)
        # При помощи исходной функции @__one_cycle_f ищем вход, соответствующий выходу @out_part
        def is_found(possible_part):
            (out_part, next_reg) = self.one_cycle_f(possible_part, reg)
            return out_part == target_out_part
        part = filter(is_found, possible_parts)
        return part[0]
    
    """
    Обобщенная функция декодирования свёрточного кода
    @encoded_bit_seq - закодированная последовательность
    @depth - глубина (де)кодирования - кол-во итераций, после которых происходит сброс внутреннего состояния
    @path_filter - функция фильтрация списка возможных путей на каждом шаге
    @path_picker - функция избирающая результирующий/наилучший путь из списка возможных путей на последней итерации
    return - исходная последовательность
    """
    def decode(self, encoded_bit_seq, depth=8, path_filter=greedy_path_filter, path_picker=arbitrary_one):
        if ((encoded_bit_seq is None) or (len(encoded_bit_seq) == 0)): return numpy.array([], dtype=int)
        # Длина закодированной подпоследовательности бит
        out_part_len = len(self.__gs)
        # Кол-во элементарных подпоследовтельностей в @encoded_bit_seq
        count = len(encoded_bit_seq)//out_part_len
        # Элементарные подпоследовательности @encoded_bit_seq
        encoded_bit_seq_parts = encoded_bit_seq.reshape(count, out_part_len)
        # Первые/корневые элементы пути кодера
        root_path_heads = self.create_possible_path_elements(None, encoded_bit_seq_parts[0], depth)
        # Текущий список вершин различных возможных путей
        path_heads = root_path_heads
        for i in xrange(1, count):
            # Элементарная подпоследовательность из @bit_seq
            encoded_part = encoded_bit_seq_parts[i]
            # Вершины всевозможных путей кодера для следующего шага
            next_path_heads = []
            for path_head in path_heads:
                next_path_heads.extend(self.create_possible_path_elements(path_head, encoded_part, depth))
            # Если достигнута глубина декодирования или последняя итерация
            if ((i > 0) and ((i%depth == 0) or (i == count-1))):
                # Выбор наилучших путей по жадному алгоритму
                path_heads = greedy_path_filter(next_path_heads)
                # Если параметр задан, то кол-во избранных путей дополнительно ограничивается
                if (MAX_SURVIVED_PATH_COUNT > 0): path_heads = arbitrary_elements(path_heads, MAX_SURVIVED_PATH_COUNT)
            else:
                # На следующий шаг "уходят" пути прошедшие фильтр, так называемые - "выжившие"
                path_heads = path_filter(next_path_heads)
        
        # Список наилуших из возможных путей должен быть не пуст
        assert len(path_heads) > 0
        
        # Выбор результирующего пути из равноценных (с точки зрения расстояния Хэмминга)
        result_path_head = path_picker(path_heads)
        
        """
        Функция декодирования пути в последовательность бит
        @path_head - вершина пути
        return - декодированная последовательность бит
        """
        def decode_path(path_head):
            decoded_signal = []
            full_path = path_head.full_path()
            for path_el in full_path:
                prev_maybe_reg = 0 if (path_el.prev() is None) else path_el.prev().maybe_reg()
                # Для декодирования на каждом элементе пути применяется обратная функция
                decoded_part_num = self.rev_one_cycle_f(path_el.maybe_out_part(), prev_maybe_reg)
                decoded_signal.extend(number_to_bit_seq(decoded_part_num, self.__k))
            return decoded_signal
        
        # Декодированные/исходные последовательности бит для каждого выигравшего/результирующего пути
        return numpy.array(decode_path(result_path_head), dtype=int)
    
    """
    Метод, производящий список возможных элеменетов пути кодера,
    происходящих из заданного элемента пути @prev.
    @prev - предыдущий элемент пути кодера
    @out_part - фактическое значение на выходе кодера (как подпоследовательность бит)
    @depth - 
    """
    def create_possible_path_elements(self, prev, out_part, depth):
        # Все возможные значения входной элементарной подпоследовательности
        possible_parts = xrange(1<<self.__k)
        # Возможное состояние кодера на предыдущем шаге @prev
        prev_maybe_reg = 0 if (prev is None) else prev.maybe_reg()
        # Функция создания возможного элемента пути
        def create_possible_path_el(maybe_part):
            (maybe_out_part, maybe_reg) = self.one_cycle_f(maybe_part, prev_maybe_reg)
            cur_len = 1 if (prev is None) else len(prev)+1
            return CC.EncoderPathEl(prev, out_part, maybe_out_part,
                                    0 if (cur_len%depth == 0) else maybe_reg)
        return map(create_possible_path_el, possible_parts)
    
    """
    Элемент пути кодера. Отражает возможное состояние кодера, на соответстсвующем шаге кодирования.
    Используется при декодировании для фиксации информации о возможных путях кодирования и выборе
    наиболее правдоподобного пути. Для этого, в том числе, сожержит ссылку на предыдущий элемент пути,
    образую тем самым однонаправленный список (путь); и текущее расстояние Хэмминга между возможной
    последовательностью, представленной данным элементом, и принятой на данном шаге последовательностью.
    """
    class EncoderPathEl:
        """
        Конструктор элемента пути
        @prev - предыдущий элемент пути
        @out_part - фактическое значение на выходе кодера (как подпоследовательность бит)
        @maybe_out_part - возможное значение на выходе кодера (как подпоследовательность бит)
        @maybe_reg - возможное состояние кодера на данном шаге
        """
        def __init__(self, prev, out_part, maybe_out_part, maybe_reg):
            self.__prev = prev
            self.__out_part = out_part
            self.__maybe_out_part = maybe_out_part
            self.__maybe_reg = maybe_reg
            # Расстояние Хэмминга. См метод @dist
            self.__dist = None
            # Длина пути до данного элемента в битах. См метод @__len__
            self.__len = None
            
        def prev(self):
            return self.__prev
        def out_part(self):
            return self.__out_part
        def maybe_out_part(self):
            return self.__maybe_out_part
        def maybe_reg(self):
            return self.__maybe_reg
        
        # Возможное значение на выходе кодера (как целое число)
        def maybe_out_part_num(self):
            return bit_seq_to_number(self.__maybe_out_part)
        
        """
        Расстояние Хэмминга (кол-во отличающихся бит) между фактическим
        и возможным значениям на выходе кодера для всей последовательности
        до текущего элемента пути.
        Результат работы функции кешируются в поле.
        """
        def dist(self):
            if (not(self.__dist is None)):
                return self.__dist
            self.__dist = numpy.sum(numpy.bitwise_xor(self.__out_part, self.__maybe_out_part))
            if (not(self.__prev is None)):
                self.__dist = self.__dist + self.__prev.dist()
            return self.__dist
        
        """
        Длина пути вплоть до данного элемента
        """
        def __len__(self):
            if (not(self.__len is None)):
                return self.__len
            self.__len = 1
            if (not(self.__prev is None)):
                self.__len = self.__len + len(self.__prev)
            return self.__len
        
        """
        Возвращает список всех элементов пути вплоть до данного
        """
        def full_path(self):
            path = []
            current_path_el = self
            while (not(current_path_el is None)):
                path.insert(0, current_path_el)
                current_path_el = current_path_el.prev()
            return path
    
    """
    Вывод общей информации о данной разновидности сверточного кода в строку
    """
    def __str__(self):
        return "CC{}/{}({})".format(self.__k, len(self.__gs), self.__gs)
```

#  Декодер Витерби

*"Жадный"* алгоритм декодирования (алгоритм по умолчанию) представленный выше неустойчив к ошибкам. Далее будет предложена реализация декодера *Витерби* на базе обобщенного декодера.

Отличие алгоритма *Витерби* от *"жадного"* алгоритма заключается в том, что на каждом шаге декодирования *"выжившими"* считаются наилучшие, для каждого из возможных состояний кодера, пути. А не наилучшие из всевозможных состояний кодера пути, как это происходит в случае *"жадного"* алгоритма.

Это позволяет повысить эффективность алгоритма в условиях ошибок, и сохранить приемлемые требования к ресурсам.

```python
"""
Функция фильтрации списка вершин вероятных путей по алгоритму Витерби.
@path_heads - вершины вероятных путей
return - список вершин "выживших" путей
"""
def viterbi_path_filter(path_heads):
    if ((path_heads is None) or (len(path_heads) == 0)): return numpy.array([], dtype=object)
    # Ширина выхода кодера
    out_width = len(path_heads[0].out_part())
    # Список групп путей на каждое возможное состояние выхода кодера
    groups = [[] for out_part_num in xrange(1<<out_width)]
    # Заполнение списка групп
    for path_head in path_heads:
        groups[path_head.maybe_out_part_num()].append(path_head)
    # Список наилучших путей из каждой группы
    result = numpy.concatenate(map(greedy_path_filter, groups))
    return result
```

# Тест декодера Витерби

### Проверка декодера Витерби на корректность декодирования сигнала без ошибок

```python
# Длина сигнала
N = 512 

# Случайный сигнал
signal = numpy.random.random_integers(0, 1, N)

# Кодирование
encoded_signal = CC_2_3.encode(signal)

# Декодирование
decoded_signal = CC_2_3.decode(encoded_signal, path_filter=viterbi_path_filter)

assert numpy.array_equal(decoded_signal, signal)

print "OK"
```

### Проверка декодера Витерби на корректность декодирования сигнала с ошибками

```python
# Длина сигнала
N = 512
# Глубина (де)кодирования
DEPTH = 64
# Кол-во ошибочных бит
ERROR_BIT_COUNT = 4

print "Глубина (де)кодирования = {}".format(DEPTH)
print "Кол-во ошибок = {}".format(ERROR_BIT_COUNT)

# Случайный сигнал
signal = numpy.random.random_integers(0, 1, N)

# Кодирование
encoded_signal = CC_2_3.encode(signal, DEPTH)

# Оптимальное расстояние между ошибками. Исключаем пакетные ошибки
error_dist = len(encoded_signal)//(2*ERROR_BIT_COUNT)
# Номера ошибочных бит
error_bit_indicies = map(lambda index: max(0, min(len(encoded_signal)-1, int(index + numpy.random.randint(2*error_dist)-error_dist))),
    numpy.linspace(0, len(encoded_signal), num=ERROR_BIT_COUNT))
# Генерация вектора ошибок
error = numpy.zeros(len(encoded_signal), dtype=numpy.int)
for error_bit_index in error_bit_indicies: error[error_bit_index] = 1
# "Внедрение" ошибок в закодированный сигнал
signal_with_error = numpy.bitwise_xor(encoded_signal, error)

# Декодирование
decoded_signal = CC_2_3.decode(signal_with_error, DEPTH, path_filter=viterbi_path_filter)

print "Кол-во ошибок после декодирования = {}".format(sum(numpy.bitwise_xor(signal, decoded_signal)))
```

# Заключение

Необходимость применения *циклических кодов* не вызывает сомнения. Задача выявления ошибочных пакетов может быть эффективно реализована при помощи *CRC*. При этом характер данных и размер пакета могут варьироваться и, по большому счёту, ничем не ограничены.

Свёрточное кодирование может быть использовано на канальном уровне для более эффективного использования спектра. Но корректирующие свойства свёрточного кода находятся на низком уровне. Хорошо подходит для случая единичных ошибок.
Но с ростом числа ошибок в канале, при декодировании происходит распространение ошибки на глубину декодирования, что приводит к образованию пакета ошибок и, как правило, увеличению их кол-ва.
Данная проблема наблюдается и в идеализированных условиях с разреженным вектором ошибок (см. последний тест).

Минимальное значение глубины зависит только от параметра скорости алгоритма (R). Максимальное - определяется размером пакета и требованиям к сложности алгоритма (производительность).

Требуется доисследование других разновидностей свёртоных кодов с низкими скоростями (низкая избыточность). С точки зрения корректировки сигнала интерес также представляет блочное линейное кодирование, но сложность реализация и высокие требования к ресурсам снижают привлекательность блочных кодов.
